```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Figure/plot title')
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--')

plt.subplot(1,3,2)
plt.plot(y,x,'g*-')

plt.subplot(1,3,3)
plt.plot(x,y,'r')
plt.plot(y,x,'g')

plt.tight_layout()
```

```python
fig1 = plt.figure()

axes = fig1.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,'b')

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title("title")
```

```python
fig2 = plt.figure()

axes_main = fig2.add_axes([0.1,0.1,0.9,0.9])

axes_inset = fig2.add_axes([0.3,0.5,0.3,0.4])

axes_main.plot(x,y,'b')
axes_inset.plot(y,x,'r')
```

```python
fig, axes = plt.subplots()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True)
plt.tight_layout()
```

```python
fig1, axes1 = plt.subplots(nrows=2, ncols=1)
plt.tight_layout()
fig2, axes2 = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
```

```python
for ax in axes1:
    ax.plot(x,y,'b')
    ax.set_xlabel('X Label')
    ax.set_ylabel('Y Label')
    ax.set_title('Title')
    
fig1.tight_layout()
fig1
```

```python
axes2[0].plot(x,y,'r')
axes2[1].plot(y,x,'b')
fig2.tight_layout()
fig2
```
